`Objetivo`: apresentar um panorama consolidado sobre a situação dos clientes no banco para o restante da empresa.

````python
# Importando a biblioteca

import pandas as pd
````

````python
# Lendo dados tabulares

credit = pd.read_excel("/content/clientes_credito.xlsx")
````

Para selecionar os valores de uma determinada coluna, podemos utilizar o operador `[]` (operador de indexação) como mostrado abaixo:

````python
# Acessando UMA coluna

credit['valor_credito']
````

A mesma operação pode ser feita com o uso do operador` .` como mostrado abaixo:

````python
credit.valor_credito 
````

Múltiplas colunas podem ser selecionadas. Podemos usar uma lista para especificar os rótulos das colunas desejada na seleção:

````python
# Acessando MÚLTIPLAS colunas

credit[['valor_pag_abr_2005','valor_pag_mai_2005','valor_pag_jun_2005','valor_pag_jul_2005','valor_pag_ago_2005','valor_pag_set_2005']]
````

Podemos usar o método `filter` para selecionar colunas baseada em seus nomes ou rótulos:

````python
# Filtrando colunas

credit.filter(like='set')
````

Podemos utilizar as propriedades `iloc` e `loc` para acessar grupos de linhas e colunas:

````python
# Acessando UMA linha

credit.loc[0]
````

````python
# Acessando MÚLTIPLAS linhas

credit.loc[0:2]
````

````python
# Acessando linhas E colunas

credit.loc[ 0:30,  [ 'ID', 'valor_credito', 'sexo', 'nivel_escolaridade' ] ]
````

````python
credit.iloc[ 0:4,  0:4 ]
````

A propriedade `loc` utiliza os índices ou rótulos dos eixos das linhas e colunas como forma de endereçamento dos elementos. Já o operador `iloc` utiliza números inteiros correspondentes à posição dos eixos (linhas e colunas).

Podemos também selecionar valores das linhas utilizando operadores lógicos: 

````python
# Acessar linhas a partir de máscaras booleanas

credit[credit["nivel_escolaridade"] == 2]
````

A condição dentro dos colchetes de seleção `credit["nivel_escolaridade"] == 2` verifica para quais linhas a coluna `nivel_escolaridade` tem um valor igual a 2, ou seja, clientes que apresentam nível de escolaridade equivalente à graduação.
