# **Tipos de dados**

## ***O que são os dados classificados como int?***

Em matemática, os inteiros são todos os números positivos ou negativos que não apresentam casas decimais. Na linguagem python os tipos de dados int representa os inteiros da matemática.
Exemplo matemático: {... -2, -1, 0, 1, 2 ...}


```python
x = 2
print(type(x))
#<class 'int'>
```

## ***O que são os dados classificados como Float?***

Em matemática, os reais são todos os números que podem ser representados em forma de fração. Na linguagem python os tipos de dados float representa os reais da matemática.
Por exemplo: { ... -3.0, -2.5, -2.0, -1.5, -1.0, -0.5, 0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0 ...}

```python
x = 2.5
print(type(x))
#<class 'float'>
```

## ***O que são os dados classificados como String?***

São conjuntos de caracteres dispostos em uma variável onde é possível trasmitir uma mensagem.

```python
x = 'Everton'
print(type(x))
#<class 'str'>
```

## ***O que são os dados classificados como Boolean?***

Os dados booleanos podem assumir apenas dois valores: False ou True.

```python
x = 1
y = 2
print(type(x == y))
#<class 'bool'>
```

## ***Estrutura da máscara***

Exemplo: {0:10d}
0 —> 1º variável no format
: —> separador
10 —> separar 10 espaços para o número
d —> número inteiro

Exemplo: {1:10.4f}
1 —> 2º variável no format
: —> separador
10 —> separar 10 espaços para o número
. —> separador decimal
4f —> limitação da quantidade das casas decimais

```python
a = 245
b = 2.835895

print('| {0:10d} | e | {1:10.4f} |'.format(a,b))
#|        245 | e |     2.8359 |
```

## ***For***

O for é uma estrutura de repetição com a seguinte sintaxe:

for i in range(inicio, fim, passo)

- **Range**
    - Início: Número inteiro inicial
    - Fim: Número inteiro final
    - Passo: Quantia somada em cada passo
- **Exemplo Range:**
    
    Range(1, 7, 1) —> 1, 2, 3, 4, 5, 6, 7
    
    Range(10, 16, 2) —> 10, 12, 14, 16
    
    Range(5, 1, -1) —> 5, 4, 3, 2, 1

**for** —> Inicio do loop

**i** —> Variável de controle

**range(1, 20)** —> variação do i

### ***Cálculos incrementais***

São cálculos realizados por atualização, segue a sintaxe:

- **Soma**
    
    soma = soma + i ou soma += i
    
- **Subtração**
    
    subt = subt - i ou subt -= i
    
- **Multiplicação**
    
    mult = mult * i ou mult *= i

## ***While***

O While é uma estrutura de repetição com a seguinte sintaxe:

while condição

While —> Início do looping

Condição —> Condição para sair do looping

### ***Exemplos***

```python
fim = 'n'
while fim == 'n':
    print('ler pagina')
    fim = input('Terminou? ')
print('Fim')
```

```python
soma = 0
while soma < 20:
    n = float(input('Digite um número: '))
    soma += n
print('A soma final foi de {:.2f}'.format(soma))
```

```python
soma = 0
while True:
    n = float(input('Digite um número: '))
    soma += n
    continua = input('Quer continuar? (s/n) ')
    if continua == 'n':
        break
print('A soma final foi de {:.2f}'.format(soma))
```

```python
dic = {'Nome': 'Everton',
        'Idade': 29,
        'Cidade': 'São José',}

print(dic['Nome'])
#Saída Everton
print(dic['Idade'])
#Saída 29
print(dic['Cidade'])
#Saída São José
print(dic)
#Saída {'Nome': 'Everton', 'Idade': 29, 'Cidade': 'São José'}
print(dic.values())
#Saída dict_values(['Everton', 29, 'São José'])
print(dic.keys())
#Saída dict_keys(['Nome', 'Idade', 'Cidade'])
print(dic.items())
#Saída dict_items([('Nome', 'Everton'), ('Idade', 29), ('Cidade', 'São José')])
```

## ***Criando nova chave***

```python
dic = {'Nome': 'Everton',
        'Idade': 29,
        'Cidade': 'São José',}

#Modidicando valor da chave Nome
dic['Nome'] = 'Tiago'

#Criando uma nova chave com um valor correspondente
dic['Peso'] = 65

print(dic['Nome'])
#Saída Tiago
print(dic['Idade'])
#Saída 29
print(dic['Cidade'])
#Saída São José
print(dic['Peso'])
#Saída 65
```

## ***Deletando chaves***

```python
dic = {'Nome': 'Everton',
        'Idade': 29,
        'Cidade': 'São José',}

del dic['Cidade']

print(dic)

#Saída {'Nome': 'Everton', 'Idade': 29}
```

## ***Dicionários dentro do for***

```python
dic = {'Nome': 'Everton',
        'Idade': 29,
        'Cidade': 'São José',}

for v in dic.values():
    print(f'Values:{v}')

#Saída
#Values:Everton
#Values:29
#Values:São José
```

```python
dic = {'Nome': 'Everton',
        'Idade': 29,
        'Cidade': 'São José',}

for k in dic.keys():
    print(f'Keys:{k}')

#Saída
# Keys: Nome
# Keys: Idade
# Keys: Cidade
```

```python
dic = {'Nome': 'Everton',
        'Idade': 29,
        'Cidade': 'São José',}

for k, v in dic.items():
    print(f'K:{k} e v:{v}')

#Saída
# K:Nome e v:Everton
# K:Idade e v:29
# K:Cidade e v:São José
```

## ***Dicionários dentro de listas***

```python
pessoas = [{'Nome': 'Everton','Idade': 29,'Cidade': 'São José'},
{'Nome': 'Thiago','Idade': 34,'Cidade': 'São Paulo'},
{'Nome':'André','Idade': 22,'Cidade': 'Lorena'}]


for p in pessoas:
    for k, v in p.items():
        print(f'Key: {k}, Value: {v}')

# Saída
# Key: Nome, Value: Everton
# Key: Idade, Value: 29
# Key: Cidade, Value: São José
# Key: Nome, Value: Thiago
# Key: Idade, Value: 34
# Key: Cidade, Value: São Paulo
# Key: Nome, Value: André
# Key: Idade, Value: 22
# Key: Cidade, Value: Lorena
```

## ***Documentando a própria função***

```python
def soma(a, b):
    """
    Função soma: soma 2 num
    a: parâmetro opcional de entrada
    b: parâmetro opcional de entrada
    retorn: s
    Autor: Everton
    """
    s = a + b
    return s

help(soma)
#Saída

#Help on function soma in module __main__:
# 
# soma(a, b)
#     Função soma: soma 2 num
#     a: parâmetro opcional de entrada
#     b: parâmetro opcional de entrada
#     retorn: s
#     Autor: Everton
```
